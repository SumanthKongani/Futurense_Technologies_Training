```python
class my_server: # every class inherits object
    def __init__(self, ip, pwd):
        self.ip = ip
        self.pwd = pwd
    def server_login(self):
        print("welcome to python", self.ip)
    
    def hostname(self):
        print("welcome to OOPs", self.ip, self.pwd)

my = my_server('2.2.2.2', '@123')
# my_server() is called a constructor
# my is now instance object reference
#when we create constructor, we will have one hidden object created inside the constructor 
#__init__ by nature is called as attributes or magic methods or the dunder methods
# __init__ will be create automatically when class constructor got created 
# __init__ as functionality we call as instance + initializer == instantiator
#__new__

my.server_login()
my.hostname()
```
